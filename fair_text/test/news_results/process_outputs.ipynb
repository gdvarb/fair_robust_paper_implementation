```python
import pandas as pd
import numpy as np
```

```python
columns=[
    "clean_error", "adv_error", "avg_bndy_error",
    "worst_clean_error", "worst_adv_error", "worst_bndy_error",
    "class0_clean_error", "class0_adv_error", "class0_bndy_error",
    "class1_clean_error", "class1_adv_error", "class1_bndy_error", 
    "class2_clean_error", "class2_adv_error", "class2_bndy_error",
    "class3_clean_error", "class3_adv_error", "class3_bndy_error"
]
```

```python
df = pd.read_csv('results_frl_distilbert_test.txt', header=None, sep='\s+', names=columns)
```

```python
df
```

```python
df['epoch'] = range(1, len(df) + 1)

std_error_avg = df[["clean_error"]].std().values[0] / np.sqrt(len(df))
std_error_worst = df[["worst_clean_error"]].std().values[0] / np.sqrt(len(df))
summary_df = pd.DataFrame({
    'epoch': df['epoch'],
    'standard_error': std_error_avg,
    'worst_standard_error': std_error_worst,  
    'avg_boundary_error': df['avg_bndy_error'],
    'worst_boundary_error': df['worst_bndy_error'],
    'avg_robust_error': df['adv_error'],
    'worst_robust_error': df['worst_adv_error']
})
```

```python
errors_df = summary_df
```

```python
columns_to_process = ['standard_error', 'worst_standard_error']
for column in columns_to_process:
    mask = errors_df.duplicated(subset=[column], keep='first')
    errors_df[column] = errors_df[column].mask(mask, np.nan)
```

```python
errors_df
```

```python
df.to_csv('labeled_results_frl_distilbert_test.csv', index=False)
errors_df.to_csv('errors_results_frl_distilbert_test.csv', index=False)
```
